```python
import pandas as pd
import numpy as np
```

```python
final_rating= pd.read_csv('final_rating.csv',encoding='latin-1')
final_rating.head()
```

```python
final_rating.shape
```

```python
book_pivot=final_rating.pivot_table(columns='user_id',index='book_title',values='num_rating')
book_pivot
```

```python
book_pivot.fillna(0,inplace=True)
book_pivot
```

```python
from scipy.sparse import csr_matrix
book_sparse=csr_matrix(book_pivot)
book_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(algorithm='brute')
model.fit(book_sparse)
```

```python
distance,sugested=model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1),n_neighbors=6)
```

```python
def recommand_book(book_name):
    book_id=np.where(book_pivot.index==book_name)[0][0]
    distance,sugested=model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1),n_neighbors=6)

    for i in range(len(sugested)):
        books=book_pivot.index[sugested[i]]
        for j in books:
            print(j)
```

```python
book_name='2nd Chance'
recommand_book(book_name)
```

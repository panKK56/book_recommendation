# **Now let's take ratings_data dataset**

```python
rating_data.head()
```

```python
# Checking which user gives how many time rating
rating_data['user_id'].value_counts()
```

```python
# Users with less rating
rating_data['user_id'].unique().shape
```

```python
# Users with high rating
x=rating_data['user_id'].value_counts()>200
x[x].shape
```

```python
y=x[x].index
y
```

```python
# USers that rated more than 200
rating_data=rating_data[rating_data['user_id'].isin(y)]
rating_data
```

```python
book_data.head(2)
```

```python
# Finding books that rated more than 200
raatings_with_books=rating_data.merge(book_data,on='isbn')
raatings_with_books
```

```python
# Finding which book got how many times of rating
num_rating=raatings_with_books.groupby('book_title')['book_rating'].count().reset_index()
num_rating
```

```python
num_rating.rename(columns={'book_rating':'num_rating'},inplace=True)
num_rating.head()
```

```python
final_rating=raatings_with_books.merge(num_rating,on='book_title')
final_rating
```

```python
# Filtering books having ratings more than 50
final_rating=final_rating[final_rating['num_rating']>=50]
final_rating.head()
```

```python
final_rating.shape
```

```python
final_rating[['book_title','user_id']].duplicated().sum()
```

```python
final_rating.drop_duplicates(['book_title','user_id'],inplace=True)
```

```python
final_rating.shape
```

```python
final_rating.to_csv('final_rating.csv',index=False)
```

# **Pivot Table**

```python
book_pivot=final_rating.pivot_table(columns='user_id',index='book_title',values='num_rating')
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0,inplace=True)
book_pivot
```
